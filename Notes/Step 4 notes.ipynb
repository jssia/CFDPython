Burgers' equation:

$$\frac{\partial u}{\partial t} + u \frac{\partial u}{\partial x} = \nu \frac{\partial^2 u}{\partial x^2}$$

Analytical solution:

\begin{eqnarray}
u &=& -\frac{2 \nu}{\phi} \frac{\partial \phi}{\partial x} + 4 \\\
\phi &=& \exp \bigg(\frac{-(x-4t)^2}{4 \nu (t+1)} \bigg) + \exp \bigg(\frac{-(x-4t -2 \pi)^2}{4 \nu(t+1)} \bigg)
\end{eqnarray}

Computational scheme:

$$u_i^{n+1} = u_i^n - u_i^n \frac{\Delta t}{\Delta x} (u_i^n - u_{i-1}^n) + \nu \frac{\Delta t}{\Delta x^2}(u_{i+1}^n - 2u_i^n + u_{i-1}^n)$$

```python
import numpy as np
import sympy
from matplotlib import pyplot as plt
from sympy import init_printing
from sympy.utilities.lambdify import lambdify
from time import sleep
%matplotlib inline

init_printing(use_latex=True)
```

```python
# Set up analytical solution
x, nu, t = sympy.symbols('x nu t')
phi = (sympy.exp(-(x - 4 * t)**2 / (4 * nu * (t + 1))) +
       sympy.exp(-(x - 4 * t - 2 * np.pi)**2 / (4 * nu * (t + 1))))
phiprime = phi.diff(x)

u = -2 * nu * (phiprime / phi) + 4

# Analytical function
uFunc = lambdify((t, x, nu), u)
```

Note boundary condition:

$$u(0) = u(2\pi)$$

```python
# Set up computational solution

## Variable declarations
l = 2 * np.pi
nx = 101
nt = 100
dx = l / (nx - 1)
nu = 0.07
dt = dx * nu

## Define domain
x = np.linspace(0, l, nx)             # Split length l into nx spacesteps
un = np.empty(nx)                     # Create array for next timestep

## Initial conditions
u = np.asarray([uFunc(0, x0, nu) for x0 in x])
```

```python
stepsize = 100
t = 1000

for step in range(stepsize, t, stepsize):
    for n in range(step):
        un = u.copy()
        for i in range(nx - 1):
            u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i - 1]) + nu * dt / dx**2 * (un[i + 1] - 2 * un[i] + un[i - 1])
        u[0] = un[0] - un[0] * dt / dx * (un[0] - un[-2]) + nu * dt / dx**2 * (un[1] - 2 * un[0] + un[-2])
        u[-1] = u[0]

    uAnalytical = np.asarray([uFunc(190 * dt, x0, nu) for x0 in x])

    plt.figure(figsize=(11, 7), dpi=100)
    plt.plot(x,u, marker='o', lw=2, label='Computational')
    plt.plot(x, uAnalytical, label='Analytical')
    plt.xlim([0, 2 * np.pi])
    plt.ylim([0, 10])
    plt.legend()
```
